```python
import numpy as np
```

```python
arr = np.array([1,2,3,4])
arr.dtype
```

```python
arr = np.array([1,2,3,4], dtype='float64')
arr.dtype
```

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.bool_)
arr
```

```python
# cast int64 to string
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.bytes_)
arr
```

```python
# cast string to int8
arr = np.array(['0','1','2','3','4'])
arr = arr.astype(np.int8)
arr
```
