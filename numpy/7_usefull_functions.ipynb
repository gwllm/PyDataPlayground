```python
import numpy as np
```

```python
arr= np.random.randint(1,20,10)
arr
```

```python
matrix= arr.reshape(2,5)
matrix
```

```python
print("max", arr.max())
print("min", arr.min())
```

```python
print('max in rows', matrix.max(1))
print('max in cols', matrix.max(0))

print('min in rows', matrix.min(1))
print('min in cols', matrix.min(0))
```

```python
# index for the max value on the whole array
print("argmax",arr.argmax())
print("argmin",arr.argmin())
```

```python
print(matrix.argmax(0))
print(matrix.argmin(0))
```

```python
# peak to peak
print(arr)
print(arr.max(),'-' ,arr.min())

print(np.ptp(arr))
```

```python
print(matrix)
print(matrix.max(0),'-', matrix.min(0))
np.ptp(matrix,0)
```
